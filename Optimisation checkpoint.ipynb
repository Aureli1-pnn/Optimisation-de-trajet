# Gestion des trajets

## Obtention des trajets

```python
# Obtention de tous les cycles domicile -> domicile passant par travail
# de longueur < 8

domicile = "Appartement Nantes"
travail  = "Travail Rennes"

def get_trajets(Graph):
    """
    Fonction permettant d'obtenir tous les cycles depuis domicile passant par travail
    """
    trajets = []
    trajet_aller  = Graph.all_paths(domicile, travail, use_multiedges=True, report_edges=True, labels=True)
    trajet_retour = Graph.all_paths(travail, domicile, use_multiedges=True, report_edges=True, labels=True)
    for trajet_a in trajet_aller:
        for trajet_b in trajet_retour:
            trajet = []
            # Deplacement représente un changement de noeud + sa valuation 
            # Exemple : Gare de Nantes -> Gare de Rennes, 550km
            for deplacement in trajet_a:
                trajet.append(deplacement)
            for deplacement in trajet_b:
                trajet.append(deplacement)
            if len(trajet) < 8:
                trajets.append(trajet)
    return trajets

trajets_transport = get_trajets(G4)
```

```python
def get_trajets_credibles(trajets):
    """
    Fontion permettant d'obtenir tous les cycles avec le même aller domicile -> travail
    que le retour travail -> domicile
    """
    trajets_credibles = []
    position_trajets  = []
    for pos, trajet in enumerate(trajets):
        possible = True
        trajet_length = len(trajet)
        # Vérification de la symétrie du trajet
        for i in range(0, int(trajet_length/2), 2):
            ville_depart_aller  = trajet[i][0]
            ville_arrive_aller  = trajet[i][1]
            ville_depart_retour = trajet[trajet_length-1-i][0]
            ville_arrive_retour = trajet[trajet_length-1-i][1]
            label_aller    = trajet[i][2]
            label_retour   = trajet[trajet_length-1-i][2]
            if ville_depart_aller!=ville_arrive_retour or ville_arrive_aller!=ville_depart_retour or label_aller!=label_retour:
                possible=False
        # Ajout du trajet s'il est symétrique
        if possible: 
            trajets_credibles.append(trajet)
            position_trajets.append(pos)
            
    return trajets_credibles, position_trajets

trajets_credibles_transport, position_trajets = get_trajets_credibles(trajets_transport)
```

```python
def get_trajets_credibles_from_graph(Graph, position_trajets):
    """
    Fait le lien entre les deux dernières fonctions pour les 
    utiliser ensemble
    """
    trajets = get_trajets(Graph)
    trajets_credibles = []
    for i, trajet in enumerate(trajets):
        if i in position_trajets:
            trajets_credibles.append(trajet)
    return trajets_credibles
```

## Affichage des trajets

```python
def show_trajet(trajets, label_name='Transport'):
    """
    Fonction affichant tous les trajets ainsi que la valuation totale de chacun d'entre eux 
    """
    for key, value in trajets.items():
        valuation = []
        
        print(str(key))
        
        # Affichage du trajet noeud par noeud 
        for i, deplacement in enumerate(value['trajet']):
            if i < len(value['trajet'])-1 : print(deplacement[0] + " --> ", end='')
            else:                  print(deplacement[0] + " --> " + deplacement[1])
    
        # Affichage des valuations de chaque arêtes empruntées 
        for j, elem in enumerate(value['all_valuation']):
            print(elem, end=" ")
            
            if label_name == 'Temps': print("min ", end="")
            elif label_name == 'Distance': print("km ", end="")
            elif label_name == 'Emission': print("KgCo2 ", end="")
            elif label_name == "Prix": print("€ ", end="")      
            if j < len(value['trajet'])-1 : print("--> ", end='')
            else: print()
                    
        # Affichage de la valuation totale du trajet
        else:
            if label_name != 'Transport':
                print(label_name + ' total: ' + str(value['total']), end=' ')
                if label_name == 'Temps': print("minutes/jour")
                elif label_name == 'Distance': print("km/semaine")
                elif label_name == 'Emission': print("Kg.Co2/an")
                elif label_name == "Prix": print("€/mois")
            else: 
                print("Nombre de changement durant le trajet : " + str(value['total']))
                
        print()
```

## Création des classements 

```python
def create_sorted_dict(trajets, label_name='Transport'):
    """
    Fonction créant un dictionnaire de trajets trié à partir 
    de la valuation totale
    """
    sorted_dict = {}
    for j, trajet in enumerate(trajets):
        all_valuation = []
        total = 0
        for deplacement in trajet:
            if label_name != 'Transport': 
                all_valuation.append(deplacement[2])
                total += deplacement[2]
            else: all_valuation.append(deplacement[2])
        
        if label_name != 'Transport': sorted_dict['Trajet ' + str(j)] = {'trajet':trajet, 'all_valuation':all_valuation, 'total':total}
        else:                         sorted_dict['Trajet ' + str(j)] = {'trajet':trajet, 'all_valuation':all_valuation, 'total':len(all_valuation)}

    return dict(sorted(sorted_dict.items(), key=lambda t: t[1]['total']))
```

```python
def afficher_classement(dict_trajet, valuation="None"):
    """
    Affiche le classement des trajets pour une valuation donnée
    """
    if valuation=="Temps":       print("Classement des trajets selon le temps de trajet par jour")
    elif valuation=="Distance":  print("Classement des trajets selon la distance parcourue par semaine")
    elif valuation=="Emission":  print("Classement des trajets selon les émissions de CO2 par an")
    elif valuation=="Transport": print("Classement des trajets selon le nombre de changement par jour")
    elif valuation=="Prix":      print("Classement des trajets selon le coût par mois")
    else:                        print("Affichage du classement")
        
    position=1
    score_precedent=-1
    
    for key, value in dict_trajet.items():
        number = [int(temp) for temp in key.split() if temp.isdigit()][0]
        if position == 1:
            print("\n1er : " + str(number), end="")
            score_precedent = value['total']
        else:
            if value['total'] <= score_precedent:
                print("-" + str(number), end="")
            else:
                print(" = " + str(score_precedent), end="")
                if valuation=="Temps":       print("min")
                elif valuation=="Distance":  print("km")
                elif valuation=="Emission":  print(" kgCo2")
                elif valuation=="Transport": print(" changements")
                elif valuation=="Prix":      print("€")
                print(str(position) + "ème : " + str(number), end="")
            score_precedent = value['total']
        position+=1        
    print(" = " + str(score_precedent), end="")
    if valuation=="Temps":       print("min")
    elif valuation=="Distance":  print("km")
    elif valuation=="Emission":  print(" kgCo2")
    elif valuation=="Transport": print(" changements")
    elif valuation=="Prix":      print("€")
```

```python
# Construction matrice de concordance 
def matrice_concordance_partielle(trajets, valuation='Temps'):
    """
    Construit la matrice de concordance partielle correspondant
    à la valuation passée en paramètre
    """
    # Récupération des seuils 
    if valuation.lower() == "temps" : 
        q = q_temps
        p = p_temps
    elif valuation.lower() == "co2" :
        q = q_emission
        p = p_emission
    elif valuation.lower() == "prix" :
        q = q_prix
        p = p_prix
    else:
        print("Veuillez rentrer une valuation valide")
        return None

    # Construction de la matrice 
    matrice = []
    for trajet_a in trajets.values():
        g_a = trajet_a[valuation.lower()]
        ligne = []
        for trajet_b in trajets.values():
            g_b = trajet_b[valuation.lower()]
            #print("Temps A : " + str(g_a) + " Temps B : " + str(g_b))
            if   g_a <= (g_b+q): c_ab = 1
            elif g_a >= (g_b+p): c_ab = 0
            else:                c_ab = ((p-(g_a-g_b))/(p-q))
            #print("Coeff " + str(c_ab))
            ligne.append(c_ab)
        matrice.append(ligne)
    
    return matrice

# Création matrice de concordance globale
def matrice_concordance_globale(m_prix, m_temps, m_co2):
    """
    Construit la matrice de concordance globale correspondant
    aux matrices passée en paramètre et aux poids défini
    plus haut 
    """
    # Création de la matrice
    matrice = []
    for i in range(len(m_prix)):
        ligne = []
        for j in range(len(m_prix)):
            c_ab = m_prix[i][j]*k_prix + m_temps[i][j]*k_temps + m_co2[i][j]*k_emission
            ligne.append(c_ab)
        matrice.append(ligne)
        
    return matrice

def affichage_matrice_concordance(matrice, valuation=None):
    """
    Affiche une matrice de concordance correspondant à la valuation
    passée en paramètre. S'il n'y a pas de valuation alors la 
    fonction affichera la matrice de concordance globale
    """
    # Affichage de l'en-tête
    if valuation!=None and valuation.lower() == "temps" : 
        print("---------------------- Matrice de concordance de Temps ----------------------")
        print("\nSeuil d'indifférence : " + str(q_temps) + "min\nSeuil de préférence  : " + str(p_temps) + "min\n")
    elif valuation!=None and valuation.lower() == "co2" : 
        print("---------------------- Matrice de concordance d'émission en CO2 ----------------------")
        print("\nSeuil d'indifférence : " + str(q_emission) + "kg\nSeuil de préférence  : " + str(p_emission) + "kg\n")
    elif valuation!=None and valuation.lower() == "prix" : 
        print("---------------------- Matrice de concordance de Prix ----------------------")
        print("\nSeuil d'indifférence : " + str(q_prix) + "€\nSeuil de préférence  : " + str(p_prix) + "€\n")
    else:
        print("---------------------- Matrice de concordance Globale ----------------------")
        print("\n             Poids - Prix : " + str(format(k_prix, ".2f")) + " - Temps : " + str(format(k_temps, ".2f")) + " - CO2 : " + str(format(k_emission, ".2f")) + "\n")
    
    # Affichage de la matrice
    i=1
    print("         Trajet 1  Trajet 2  Trajet 3  Trajet 4  Trajet 5  Trajet 6")
    print("-------------------------------------------------------------------")
    for ligne in matrice:
        print("Trajet " + str(i), end=" |  ")
        for coefficient in ligne:
            len_coeff = len(str(round(coefficient, 4)))
            end = (10-len_coeff)*" "
            print(round(coefficient, 4), end=end)
        print()
        i+=1
    print() 
```

```python
######### Création des matrices de concordances partielles #########

# Matrice de Temps
matrice_temps = matrice_concordance_partielle(mes_trajets, "temps")
affichage_matrice_concordance(matrice_temps, "temps")

# Matrice d'émission
matrice_co2 = matrice_concordance_partielle(mes_trajets, "co2")
affichage_matrice_concordance(matrice_co2, "co2")

# Matrice de Prix
matrice_prix = matrice_concordance_partielle(mes_trajets, "prix")
affichage_matrice_concordance(matrice_prix, "prix")

########### Création dela matrice de concordance globale ###########
m_concord_globale = matrice_concordance_globale(matrice_prix, matrice_temps, matrice_co2)
affichage_matrice_concordance(m_concord_globale)
```

## Matrice de discordance

### Définitions des seuils de véto

```python
# Seuil de véto 
v_prix     = 25
v_temps    = 20
v_emission = 3000
```

### Construction et affichage de la matrice de discordance

```python
# Construction matrice de discordance
def matrice_discordance(trajets):
    """
    Construit la matrice de disconcordance grâce 
    aux données comprises dans trajets et aux 
    seuils définis plus haut
    """
    matrice = []
    for trajet_a in trajets.values():
        ligne = []
        for trajet_b in trajets.values():
            prix_a     = trajet_a["prix"]
            prix_b     = trajet_b["prix"]
            emission_a = trajet_a["co2"]
            emission_b = trajet_b["co2"]
            temps_a    = trajet_a["temps"]
            temps_b    = trajet_b["temps"]
            
            if   (prix_a >= prix_b+v_prix) or (emission_a >= emission_b+v_emission) or (temps_a >= temps_b+v_temps): d_ab = 1
            else : d_ab = 0

            #print("Coeff " + str(c_ab))
            ligne.append(d_ab)
        matrice.append(ligne)
    
    return matrice

def affichage_matrice_discordance(matrice):
    """
    Affiche la matrice de discordance
    """
    # Affichage de l'en-tête
    print("---------------------- Matrice de discordance ----------------------")
    print("\n     Seuil de véto - Prix : " + str(v_prix) + "€ - Temps : " + str(v_temps) + "min - CO2 : " + str(v_emission) + "kg")
    
     # Affichage de la matrice
    i=1
    print("\n         Trajet 1  Trajet 2  Trajet 3  Trajet 4  Trajet 5  Trajet 6")
    print("---------------------------------------------------------------------")
    for ligne in matrice:
        print("Trajet " + str(i), end=" |  ")
        for coefficient in ligne:
            len_coeff = len(str(coefficient))
            end = (10-len_coeff)*" "
            print(coefficient, end=end)
        print()
        i+=1
    print() 
```

### Graphe de surclassement

```python
def graphe_surclassement(m_surclassement):
    """
    Créé le graphe de surclassement à partir 
    de la matrice du même nom.
    Retourne un objet DiGraph de SageMath
    """
    # Création du graphe
    G = DiGraph(loops=True)
    
    # Création des noeuds
    noeuds = [1, 2, 3, 4, 5, 6]
    G.add_vertices(noeuds)
    
    # Création des arêtes
    i=1
    for ligne in m_surclassement:
        j=1
        for coeff in ligne:
            if coeff==1 and j!=i: G.add_edge(i, j)
            j+=1 
        i+=1
        
    return G
```

### Supression des circuits

#### Transformation de la forme du graph    MATRICE -> LISTE 

```python
def transform_in_adjency_list(m_adjacence):
    """
    Retourne le graphe de surclassement sous forme de liste d'adjacence
    """
    list_adjacence = {}
    dim_m = len(m_surclassement)
    for i in range(dim_m):
        successeur = []
        for j in range(dim_m):
            if m_surclassement[i][j] == 1: successeur.append(j+1)
        list_adjacence[i+1] = successeur
        
    return list_adjacence
```

#### Obtention des circuits dans le graphe 

```python
def get_circuit(list_adjacence):
    """
    Retourne le premier circuit trouvé dans le Graphe
    """    
    circuits = []
    def DFS(actuel, depart, chemin):
        """
        Parcours en profondeur d'un sous-graphe
        """
        chemin.append(actuel)
        for voisin in list_adjacence[actuel]:
            if voisin not in chemin:
                DFS(voisin, depart, chemin)
            elif voisin == depart:
                chemin = sorted(chemin)
                circuits.append(chemin[:])
        chemin.pop()
    
    # Vérifie la présence de circuit pour chaque noeud du graphe
    for noeud, successeur in list_adjacence.items():
        chemin = []
        DFS(noeud, noeud, chemin)
    
    # Si au moins un circuit a été trouvé alors on retourne le plus long
    if len(circuits) > 0:
        circuits = sorted(circuits, key=lambda t: -len(t)) 
        return circuits[0]
    
    # Aucun circuit n'a été trouvé
    return None
```

#### Suppression des circuits

```python
def delete_circuit(Graph):
    """
    Supprime les circuits du Graph de surclassement
    Graph : sous forme d'une matrice d'adjacence
    Return : Version du graphe d'entrée sans circuit sous forme de list d'adjacence
             Liste de la correspondance entre les nouveaux noeuds (composé) et les anciens
    """
    # Transforme la matrice d'adjacence en liste d'adjacence
    Graph = transform_in_adjency_list(Graph)
    
    # Récupère le tout premier circuit du graph
    circuit = get_circuit(Graph)
    
    # Contracte tout les noeud du circuit en un seul tant qu'il y a encore des circuits
    correspondance = {}
    while(circuit != None):
        
        # Création du nom du nouveau noeud à partir du noms des noeuds qui le compose
        # Ex : circuit(1,2,3) -> Nom du noeud générique 123
        nom_nouveau_noeud = 0
        dizaine = 0
        for i in range(len(circuit) - 1, -1, -1):
            nom_nouveau_noeud += circuit[i]*(10**dizaine)
            dizaine+=1
        
        # Ajout du nouveau noeud dans le dictionnaire de correspondance
        correspondance[nom_nouveau_noeud] = circuit

        # Récupération des arcs avec les sommets extérieurs au circuit 
        successeurs_nouveau_noeud = []
        for noeud, successeurs in Graph.items():
            if noeud in circuit: 
                for elem in successeurs: 
                    if (elem not in successeurs_nouveau_noeud) and (elem not in circuit): 
                        successeurs_nouveau_noeud.append(elem)

        # Création du nouveau Graph
        new_graph = {}
        for noeud, successeurs in Graph.items():
            if noeud not in circuit: 
                new_successeur = []
                for elem in successeurs: 
                    if elem not in circuit: new_successeur.append(elem)
                    else: 
                        if nom_nouveau_noeud not in new_successeur: new_successeur.append(nom_nouveau_noeud)
                new_graph[noeud] = new_successeur
        new_graph[nom_nouveau_noeud] = successeurs_nouveau_noeud
        
        # Cherche un circuit dans le nouveau graph
        Graph = new_graph
        circuit = get_circuit(Graph)
        
    return Graph, correspondance
```

### Obtention du noyau du graph de surclassement

```python
def get_noyau(Graph):
    """
    Retourne le noyau sous forme d'un tableau. Il est possible qu'il contienne
    des sommets composées de plusieurs autres.
    """
    ##################### Obtention du dictionnaire des précédents ######################
    
    list_predecesseurs = {}
    for noeud in Graph.keys():
        pred_noeud = []
        for key, successeurs in Graph.items():
            if noeud != key:
                if noeud in successeurs: pred_noeud.append(key)
        list_predecesseurs[noeud] = pred_noeud
    
    ########################## Algorithme d'obtention du noyau ##########################
    noyau = []

    # Fonction permettant d'obtenir les noeuds n'ayant aucun prédecesseur
    def noeud_sans_pred(list_pred):
        noeud_marque = []
        for noeud, predecesseurs in list_predecesseurs.items():
            # On inclut pas les éléments déjà dans le noyau pour éviter d'avoir une boucle while infini
            if (len(predecesseurs) < 1) and (noeud not in noyau): 
                noeud_marque.append(noeud)
                
        return noeud_marque
    
    # Tant que possible faire
    noeud_marque = noeud_sans_pred(list_predecesseurs)
    while(len(noeud_marque) > 0):
        # Ajout des noeuds sans predecesseurs dans le noyau
        for noeud in noeud_marque: 
            noyau.append(noeud)
        # Barre les sommets ayant au moins un sommet marqué comme prédecesseur
        sommets_barre = []
        for key, predecesseurs in list_predecesseurs.items():
            if any(element in predecesseurs for element in noeud_marque): sommets_barre.append(key)
        # Supprime des listes les sommets barrés
        new_list_predecesseur = {}
        for key, predecesseurs in list_predecesseurs.items():
            # Copie la liste des prédecesseurs sans les sommets barréss
            pred_noeud = [element for element in predecesseurs if element not in sommets_barre] 
            new_list_predecesseur[key] = pred_noeud
        
        list_predecesseurs = new_list_predecesseur
        noeud_marque = noeud_sans_pred(list_predecesseurs)
    
    return noyau
```

#### Affichage du noyau

```python
def affichage_noyau(Noyau, correspondance):
    """
    Affiche les sommets faisant parti du noyau grâce à la correspondance 
    entre les sommets composés et les vrais sommets
    """
    print("Noyau : ", end="")
    for j in range(len(Noyau)):
        elem = Noyau[j]
        if elem in correspondance: 
            noeud_originaux = correspondance[elem]
            for i in range(len(noeud_originaux)):
                print(noeud_originaux[i], end="")
                if i < len(noeud_originaux)-1: print("-", end="")
        else: print(elem)
        if j < len(Noyau)-1: print("-", end="")
```
